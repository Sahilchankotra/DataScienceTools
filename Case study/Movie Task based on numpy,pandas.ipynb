```python
# Import all needed libraries
import numpy as np
import pandas as pd
```

### Reading the data 

```python
data=pd.read_csv('Movie+Assignment+Data.csv')
data.head(10)
```

### Inspect the data Frame

```python
print(data.shape)
print(data.info())
```

### Cleaning the data

```python
# Check the null values
print(data.isnull().sum())
```

```python
### Write your code for column-wise null percentages here
round(data.isnull().sum() *100/len(data),2)
```

```python
## So many of the columns in this dataframe are not required. It advised to drop columns
data=data.drop(['color','director_facebook_likes','actor_1_facebook_likes','actor_2_facebook_likes','actor_3_facebook_likes','actor_2_name','cast_total_facebook_likes','actor_3_name','duration','facenumber_in_poster','content_rating','country','movie_imdb_link','aspect_ratio','plot_keywords'],axis='columns')
data
```

```python
#  Drop unecessary rows using columns with high Null percentages
f=round(data.isnull().sum() *100/len(data),2)
h=f.values
g=np.where(h>5)
name=f.index
for i in g:
    print(name[i])
    data=data.dropna(subset=name[i])
print(data.head())
```

```python
## Now again check the null percentage
round(data.isnull().sum() *100/len(data),2)
```

```python
# You can see we have some nan values in language columns. We can fill this value with English
data['language']=data['language'].fillna('English')
print(round(data.isnull().sum()*100/len(data),2))
```

### Data Analysis

```python
# Find the movies with highest profit
Profit=data['gross']-data['budget']
data['Profit']=Profit
```

```python
# Write your code for sorting the dataframe her
data.sort_values(by='Profit',ascending=False)
```

```python
# Top 10 movies highest gross
top10=data.sort_values(by='Profit',ascending=False).head(10)
top10
```

```python
# in above top10 dataset you can see we have some duplicate values. Next task is to remove dublicate
data.drop_duplicates()
data.head(10)
```

```python
# Write your code for extracting the top 250 movies as per the IMDb score here. Make sure that you store it in a new dataframe 
# and name that dataframe as 'IMDb_Top_250'
IMDb_top_250=data[data['num_voted_users']>25000]
IMDb_top_250=IMDb_top_250.sort_values(by='imdb_score',ascending=False).head(250)
IMDb_top_250
```

```python
## Top foreign language movies
topFL=IMDb_top_250[IMDb_top_250['language']!='English']
topFL.head(20)
```

```python
# Write your code for extracting the top 10 directors here
top10dir=data.groupby('director_name').imdb_score.mean().sort_values(ascending=False).head(10)
top10dir
```

```python
##Find popular genres
genres=data['genres'].str.split("|",n=-1, expand=True)
data['genre_1']=genres[0]
data['genre_2']=genres[1]
data.head(4)
```

```python
#Most popular combo of genres
popgen=data.groupby(['genre_1','genre_2']).gross.mean().sort_values(ascending=False).head(5)
popgen
```

```python
top_actor=data.actor_1_name.value_counts().sort_values(ascending=False).head(5)
top_actor
```
