# Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.data visulizations means represent the data in a pictorial or graphical format

```python
# Import Library
import matplotlib.pyplot as plt
from matplotlib import style
```

### Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias

```python
#First program
import numpy as np
style.use("ggplot")
xpoints=np.array([0,6])
ypoints=np.array([0,250])
plt.plot(xpoints,ypoints,label="Firstline")
plt.legend()
plt.grid(True,color='k')
plt.show()
```

## The plot() function is used to draw points (markers) in a diagram.
## By default, the plot() function draws a line from point to point.

```python
### Plotting Without Line

xpoints=np.array([0,8])
ypoints=np.array([2,32])
plt.plot(xpoints,ypoints,'o')
plt.show()
```

```python
#Markers
xpoints=np.array([2,3,4,5])
ypoints=np.array([5,3,2,1])
plt.plot(xpoints,ypoints,marker='o')
plt.show()
```

```python
## Format string fmt
## use the shortcut string notation parameter to specify the marker.
## ms used for marker size
## mec used for marker color
xpoints=np.array([32,4,1,3,5])
ypoints=np.array([31,24,5,3,2])
plt.plot(xpoints,ypoints,'*-.r',ms=20,mec='b')
plt.show()
```

```python
## Display Multiple Plots
# Subplots: with the help of subplots you can draw multiple plots

x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(1,4,4)
plt.plot(x,y)
plt.show()
```

```python
x=np.array([2,3,4,5])
y=np.array([32,42,1,4])
plt.subplot(2,3,1)
plt.plot(x,y)
plt.title("Sales")

x=np.array([32,21,43,54])
y=np.array([6,5,4,3])
plt.subplot(2,3,2)
plt.plot(x,y)
plt.title("income")
plt.show()

x=np.array([11,22,33,44])
y=np.array([66,77,8,9])
plt.subplot(2,3,3)
plt.plot(x,y)
plt.title("Overall salary")
```

### Scatterplot: This  function plots one dot for each observation. It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis:

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
print(len(x))
print(len(y))
plt.scatter(x,y)
plt.xlabel("years")
plt.ylabel("KM")
plt.show()
```

```python
# Compare plots

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)


x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)

plt.show()
```

```python
# Create random array with 100 points and colors

x=np.random.randint(100,size=(100))
y=np.random.randint(100,size=(100))
colors=np.random.randint(100, size=(100))
sizes=5*np.random.randint(100, size=(100))
plt.scatter(x,y,c=colors, alpha=0.5, s=sizes, cmap='nipy_spectral')
plt.show()
```

```python
## Bars
x=np.array([2,3,4,5])
y=np.array([7,6,5,4])
plt.bar(x,y)
plt.show()
```

```python
x=np.array(['A','B','C','D'])
y=np.array([53,40,52,60])
plt.barh(x,y)
plt.show()
```

```python
## Histograms

x=np.random.normal(170,10,250)
print(x)
```

```python
plt.hist(x)
plt.show()
```

```python
# Piecharts

x=np.array([35,45,64,63])
plt.pie(x)
plt.show()
```

```python
#Labels
y=np.array([32,42,21,54,6])
la=["A","B","C","D","E"]
plt.pie(y,labels=la)
plt.show()
```

```python
# if you want one of the wedges to stand out you can use explode
y=np.array([35,45,67,88])
la=["A","B","C","D"]
ex=[0.2,0,0,0]
plt.pie(y,labels=la,explode=ex, autopct='%1.1f%%')
plt.show()
```
